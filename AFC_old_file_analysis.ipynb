```python
import pandas as pd
import numpy as np
```

```python
df_org=pd.read_excel('MiDas.xlsx')
```

```python
df_org.tail(10)
```

```python
col=df_org.columns
```

```python
col_null=[]
notnull_col=[]
col_null_any=[]
for k in col:
  if df_org[k].isnull().any() == True:
    col_null_any.append(k)
  if df_org[k].isnull().all() ==True:
    col_null.append(k)
  if df_org[k].notnull().all() == True:
    notnull_col.append(k)

print(col_null)
print(col_null_any)
print(notnull_col)
```

```python
print(len(col_null_any))
print(len(col_null))
print(len(notnull_col))
```
